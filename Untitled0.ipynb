```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import mplcursors

def calculate_returns(trading_days, per_day_risk, rrr, winning_prop):
    return trading_days * per_day_risk * (winning_prop * (rrr + 1) - 1)

def update_graph(trading_days, per_day_risk, rrr, winning_prop, varying_param):
    if varying_param == 'Trading Days':
        param_range = np.linspace(1, 30, 100)
    elif varying_param == 'Per Day Risk':
        param_range = np.linspace(0.1, 10, 100)
    elif varying_param == 'R.R.R':
        param_range = np.linspace(0.1, 10, 100)
    elif varying_param == 'Winning Prop':
        param_range = np.linspace(0.1, 1, 100)

    returns = []
    for param_value in param_range:
        if varying_param == 'Trading Days':
            returns.append(calculate_returns(param_value, per_day_risk, rrr, winning_prop))
        elif varying_param == 'Per Day Risk':
            returns.append(calculate_returns(trading_days, param_value, rrr, winning_prop))
        elif varying_param == 'R.R.R':
            returns.append(calculate_returns(trading_days, per_day_risk, param_value, winning_prop))
        elif varying_param == 'Winning Prop':
            returns.append(calculate_returns(trading_days, per_day_risk, rrr, param_value))

    plt.figure(figsize=(10, 6))
    line, = plt.plot(param_range, returns, marker='o')

    # Set more granular ticks
    if varying_param == 'Winning Prop':
        plt.xticks(np.arange(0, 1, 0.05))
    elif varying_param == 'Trading Days':
        plt.xticks(np.arange(1, 31, 1))
    elif varying_param == 'Per Day Risk' or varying_param == 'R.R.R':
        plt.xticks(np.arange(0.1, 10, 0.5))

    plt.yticks(np.arange(min(returns), max(returns) + 1, 5))
    # plt.yticks(np.arange(-50, 50, 5))

    plt.ylabel('Returns')
    plt.xlabel(varying_param)
    plt.title('Returns per Month')
    plt.grid(True)

    # Add interactive cursor
    mplcursors.cursor(line, hover=True).connect("add", lambda sel: sel.annotation.set_text(f"x: {sel.target[0]:.2f}\ny: {sel.target[1]:.2f}"))

    plt.show()

# Widgets
trading_days_slider = widgets.IntSlider(value=20, min=1, max=30, step=1, description='Trading Days:')
per_day_risk_slider = widgets.FloatSlider(value=2, min=0.1, max=10, step=0.1, description='Per Day Risk:')
rrr_slider = widgets.FloatSlider(value=3, min=0.1, max=10, step=0.1, description='R.R.R:')
winning_prop_slider = widgets.FloatSlider(value=0.25, min=0, max=1, step=0.01, description='Winning Prop:')
varying_param_dropdown = widgets.Dropdown(options=['Trading Days', 'Per Day Risk', 'R.R.R', 'Winning Prop'],
                                          value= 'Winning Prop' , description='Varying Parameter:')

interactive_plot = widgets.interactive(update_graph,
                                       trading_days=trading_days_slider,
                                       per_day_risk=per_day_risk_slider,
                                       rrr=rrr_slider,
                                       winning_prop=winning_prop_slider,
                                       varying_param=varying_param_dropdown)

display(interactive_plot)
```
